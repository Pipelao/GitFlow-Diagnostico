```python
import json
import pandas as pd
import regex as re

data = [json.loads(line) for line in open('farmers-protest-tweets-2021-03-5.json', 'r')]
```

```python
# Flatten data
df_nested_list = pd.json_normalize(data)
```

```python
df_nested_list
```

## Pregunta 1

### Los top 10 tweets m√°s retweet

```python
def pregunta_1(df):
    retweet_top = df.sort_values(by=['retweetCount'], ascending=False)
    return retweet_top.head(10)

pregunta_1(df_nested_list)
```

## Pregunta 2

### Los top 10 usuarios en funci√≥n a la cantidad de tweets que emitieron

```python
def pregunta_2(df):
    usuarios_top = df['user.username'].value_counts()
    return usuarios_top[:10]

pregunta_2(df_nested_list)
```

## Pregunta 3

### Los top 10 d√≠as donde hay m√°s tweets

```python
def pregunta_3(df):
    df_new = df.copy()
    df_new['date'] = df['date'].apply(lambda x: x[:10])
    dias_top = df_new['date'].value_counts()
    return dias_top[:10]

pregunta_3(df_nested_list)
```

## Pregunta 4

### Top 10 hashtags m√°s usados

```python
def pregunta_4(df):
    df_new = df.copy()
    df_new['hashtags'] = df['content'].apply(lambda x: re.findall(r"#(\w+)", x))
    return df_new.explode('hashtags')['hashtags'].value_counts()[:10]

pregunta_4(df_nested_list)
```

```python
def Main(df):
    respuesta_1 = pregunta_1(df)
    respuesta_2 = pregunta_2(df)
    respuesta_3 = pregunta_3(df)
    respuesta_4 = pregunta_4(df)
    print("Respuesta a preunta 1: ", respuesta_1)
    print("-------------------------------------")
    print("Respuesta a pregunta 2: ", respuesta_2)
    print("-------------------------------------")
    print("Respuesta a pregunta 3: ", respuesta_3)
    print("-------------------------------------")
    print("Respuesta a pregunta 4: ", respuesta_4)
```

```python
Main(df_nested_list)
```
